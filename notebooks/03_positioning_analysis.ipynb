# 03 · Positioning Analysis (exploration)

Exploratory notebook for **Decision Question 2** — CRM vs Revenue Operations Platform vs AI
Readiness Platform. Applies the Positioning Canvas weights (from YAML) to a **synthetic** analyst
assessment via the shared `weighted_score` function. Illustrative (grade C); positioning must be
decided on buyer evidence. Companion: `dashboards/positioning-dashboard.qmd`,
`working-paper/04-crm-positioning.qmd`.

```python
import sys, pathlib
_root = pathlib.Path.cwd()
while _root != _root.parent and not (_root / 'src').is_dir():
    _root = _root.parent
sys.path.insert(0, str(_root))

import pandas as pd
import plotly.express as px

from src.utils import load_framework, weights_from_framework
from src.scoring import weighted_score
```

```python
# Synthetic analyst assessment: each option scored 0-10 per criterion.
assessment = pd.DataFrame(
    [
        ['CRM Platform',                 9, 4, 7, 5, 6, 6],
        ['Revenue Operations Platform',  6, 7, 8, 7, 8, 5],
        ['AI Readiness Platform',        5, 8, 6, 9, 8, 3],
    ],
    columns=['option', 'clarity', 'differentiation', 'customer_pain_alignment',
             'investor_appeal', 'scalability', 'evidence_strength'],
)
assessment
```

```python
weights = weights_from_framework(load_framework('positioning_canvas'))
assessment['positioning_score'] = assessment.apply(lambda r: weighted_score(r, weights), axis=1)
ranked = assessment.sort_values('positioning_score', ascending=False).reset_index(drop=True)
ranked[['option', 'positioning_score']]
```

```python
fig = px.bar(ranked.sort_values('positioning_score'), x='positioning_score', y='option',
             orientation='h', range_x=[0, 100],
             title='Weighted positioning score (illustrative, synthetic)')
fig.show()
```
