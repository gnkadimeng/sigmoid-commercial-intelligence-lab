# 02 · AI Readiness Index (exploration)

Exploratory notebook operationalising the proposal's core claim — *the CRM is the foundation for
AI readiness* (**Hypothesis H-CRM-AI**). Uses shared logic from `src/` on **synthetic** customer
data. Scores are illustrative (grade C/D); the claim itself is not yet validated. Companion:
`dashboards/ai-readiness-scorecard.qmd`, `working-paper/06-ai-readiness.qmd`.

```python
import sys, pathlib
_root = pathlib.Path.cwd()
while _root != _root.parent and not (_root / 'src').is_dir():
    _root = _root.parent
sys.path.insert(0, str(_root))

import pandas as pd
import plotly.express as px

from src.utils import load_sample
from src.scoring import calculate_ai_readiness_score, calculate_customer_fit_score
```

```python
customers = load_sample('customer_profiles.csv').copy()
customers['ai_readiness_score'] = customers.apply(calculate_ai_readiness_score, axis=1)
customers['customer_fit_score'] = customers.apply(calculate_customer_fit_score, axis=1)
customers[['company_name', 'industry', 'ai_readiness_score', 'customer_fit_score']]\
    .sort_values('ai_readiness_score', ascending=False)
```

```python
fig = px.histogram(customers, x='ai_readiness_score', nbins=8,
                   title='AI readiness distribution (synthetic)')
fig.show()
```

```python
# Is readiness related to fit? An empirical question, not an assertion.
fig = px.scatter(customers, x='customer_fit_score', y='ai_readiness_score',
                 hover_name='company_name', size='data_quality',
                 title='AI readiness vs CRM fit (bubble = data quality)')
fig.show()
```

**Caution:** the index *encodes* the thesis (data quality heavily weighted). Validating H-CRM-AI
requires linking these scores to **actual** AI outcomes in a pilot (`EV-010`).
