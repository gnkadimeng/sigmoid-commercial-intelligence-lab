# 01 · Market Prioritisation (exploration)

Exploratory notebook for **Decision Question 1** — which segment to target first for the CRM.
It imports the shared scoring logic from `src/` (the single source of computational truth) and
works on the **synthetic** sample data. Scores are illustrative (confidence grade C), not a
recommendation. The rendered companion is `dashboards/market-prioritisation.qmd`.

```python
import sys, pathlib
# Make the project root importable regardless of where Jupyter was launched.
_root = pathlib.Path.cwd()
while _root != _root.parent and not (_root / 'src').is_dir():
    _root = _root.parent
sys.path.insert(0, str(_root))

import pandas as pd
import plotly.express as px

from src.utils import load_sample, load_framework
from src.scoring import calculate_market_priority_score
```

```python
segments = load_sample('market_segments.csv').copy()
segments['priority_score'] = segments.apply(calculate_market_priority_score, axis=1)
ranked = segments.sort_values('priority_score', ascending=False).reset_index(drop=True)
ranked[['segment_name', 'industry', 'priority_score']]
```

```python
fig = px.bar(ranked.sort_values('priority_score'), x='priority_score', y='segment_name',
             orientation='h', range_x=[0, 100],
             title='Market priority score by segment (illustrative, synthetic)')
fig.show()
```
